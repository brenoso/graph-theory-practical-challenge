### Diagrama de Voronoi - Explicação

Um Diagrama de Voronoi é uma partição de um plano em regiões com base na distância para pontos em um subconjunto específico do plano (sites). Um diagrama de Voronoi é:

*-(geralmente) Duas dimensões - esta é uma partição de um plano .
-Determinístico - dado um conjunto de pontos, você obterá o mesmo particionamento em cada execução do algoritmo de particionamento.
De granularidade variável - partes do plano com muitos pontos terão muitas células de Voronoi. Uma parte mais esparsa terá menos células.
Parâmetro livre.
Otimizando para distâncias de cada ponto no plano a partir dos locais - cada ponto no plano estará na célula de Voronoi do local mais próximo, ou na fronteira entre várias células de locais nos quais é equidistante.
Tem complexidade computacional de O (nlogn) tempo e O (n) espaço para encontrar, onde n é o número de pontos.
O resultado do algoritmo é uma lista de polígonos - a partição.*
